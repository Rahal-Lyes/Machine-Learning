```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import os
import matplotlib.pyplot as plt
```

```python
print(os.listdir())
```

```python
df=pd.read_csv('canada_per_capita_income.csv')
print(df.columns)
```

```python
df = df.rename(columns={df.columns[1]: "price"})
df.columns
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.xlabel("year")
plt.ylabel("price")
plt.scatter(df["year"], df["price"], color="red", marker="+")
plt.show()


```

```python
rg=linear_model.LinearRegression()
```

```python
rg.fit(df[['year']],df.price)
```

```python
rg.predict([[2020]])
```

```python
data2={
    "year":[2020,2021,2022,2023,2024,2025,2026]
}

df2=pd.DataFrame(data2)
```

```python
df2['price']=rg.predict(df2)
```

```python
df2
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.xlabel("year")
plt.ylabel("price")
plt.scatter(df2["year"], df2["price"], color="green", marker="+")
plt.show()
```
