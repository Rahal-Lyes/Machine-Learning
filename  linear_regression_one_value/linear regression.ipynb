```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data={
"area":[2600,3000,3200,3600,4000],
    "price":[5500,5600,6100,6800,7200]
}
```

```python
df=pd.DataFrame(data)
```

```python
%matplotlib inline
plt.xlabel("Surface")
plt.ylabel("prix")
plt.scatter(df.area,df.price,color="red",marker="+")
```

```python
model=linear_model.LinearRegression()
#training data to have best minimise erreur
model.fit(df[['area']],df.price)
# rg.fit(df[['area']],df.price)
```

```python
model.predict([[3333]])
  #price=m * area +b m==>coef_ ,intercept_=>b
```

```python
data_without_price={
    "area":[10000,20000,30000,40000,50000],
    
}
```

```python
df2=pd.DataFrame(data_without_price)
```

```python
df2
```

```python
df2['prices']=model.predict(df2)
```

```python
df2
```

```python
df2.to_csv('prediction.csv')
```

# Save Model In File

```python
import pickle
```

```python
with open('model_pickle','wb') as file:
    pickle.dump(model,file)
    #save model='model' in file 
```

```python
with open('model_pickle','rb') as file:
    model_file=pickle.load(file)
    
```

```python
model_file.predict([[3333]])
```
