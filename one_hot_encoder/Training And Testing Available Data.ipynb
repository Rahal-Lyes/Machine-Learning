```python
import pandas as pd
import os
import matplotlib.pyplot as plt
print(os.listdir())
```

```python
df=pd.read_csv('carprices_bmw.csv')
df
```

```python
%matplotlib inline
plt.xlabel('price')
plt.ylabel('kilometrage')
plt.scatter(df['Sell Price($)'],df['Mileage'])
```

```python
%matplotlib inline
plt.ylabel('price')
plt.xlabel('Age(yrs)')
plt.scatter(df['Age(yrs)'],df['Sell Price($)'])
```

```python
X=df[['Mileage','Age(yrs)']]
y=df['Sell Price($)']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(X,y,test_size=0.2)
```

```python
len(y_train)
```

```python
len(x_test)
```

```python
from sklearn.linear_model import LinearRegression
rg=LinearRegression()
```

```python
rg.fit(x_train,y_train)
```

```python
rg.predict(x_test)
```

```python
y_test
```

```python
rg.score(x_test,y_test)
```
